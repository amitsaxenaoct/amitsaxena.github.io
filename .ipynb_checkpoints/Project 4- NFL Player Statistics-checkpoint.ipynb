#### **Ethical Implications**

In my opinion using the data ethically is very important for all data scientists. 
- Issues of transparency in data collection, research, and dissemination are important points. There should never be any doubt about the ownership of personal information, and only the data which is allowed to be used should be used in analysis.
- Transparency in data collection, storage, and use should always be maintained and if any work product created by someone else is being used then we have to give credit to the person and reference the source clearly and explicitly.
- Ensuring data subjects' privacy is of utmost importance.
- Preservation of privacy where data is sensitive and ensuring data subjects' privacy should be of utmost importance.
- We should always try not to have any kind of bias in data selection, because it’ll skew the analysis and may become counter productive.
- Prevention of disruption and "hacking" of data should always be avoided by placing necessary safeguards in place. Using encryption and passwords is a good practice.
- Misuse or mishandling of data should always be avoided.
- Possibility of harm by the misuse of data is always a possibility and we must make sure that the data isn’t used in such a way that it could cause any harm to anybody.
- Preservation of privacy where data is sensitive or personal is also another important point to remember. If we’re collecting data which has some personally identifiable information then that should be removed before using it in any analysis.
